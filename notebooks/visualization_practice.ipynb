```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# group by genre and calculate the average total streams
avg_greams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_greams
```

```python
plt.bar(avg_greams.index, avg_greams.values)
plt.show()
```

```python
plt.bar(avg_greams.index, avg_greams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_greams.index, avg_greams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha ="right")

plt.tight_layout()
plt.show()
```

```python
# Eliminate borders

plt.figure(figsize=(10,6))
plt.bar(avg_greams.index, avg_greams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha ="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ["#7b92c9"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#c97ba8"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artits")
plt.xticks(rotation=45, ha ="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = "Ariana Grande"

colors = ["#c97ba8" if artist == special_artist else "grey" for artist in avg_artist_streams.index] 
 

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.9)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande rans 2nd in Total Average Streams")
plt.xticks(rotation=45, ha ="right")

plt.tight_layout()
plt.show()
```

```python

plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], spotify_df["Total Streams (Millions)"])
plt.show()
```

```python

plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"], 
color = "#15c72fff",
# transperancy 
alpha = 0.8,
s= 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
plt.title("No Appearent Correlation Between Monthly Listends and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)


plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#15c72fff" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"], 
color = colors,
# transperancy 
alpha = 0.8,
s= 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")


ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listends or Total Streams?", fontsize = 18, pad=35)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
# Line charts

yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(12, 6))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker="+",
    linestyle = "--",
    color = "#036b16",
    linewidth = 0.5)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad=20)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color="#063b0a",
    linewidth = 0.5,
    alpha = .07)

ax.text(
    2022+0.05,
    2000,
    "Dip In Total Streams",
    color = "#063b0a",
    fontsize=8
)

plt.tight_layout()

plt.ylim(0, 3500)
plt.show()
```

```python
# Pie charts

category = "Platform Type" 
values = "Total Streams (Millions)" 

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=None,
    startangle=110,
    colors=["grey", "#1aaa10"],
autopct="%1.1f%%")

plt.title("Near Even Split Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index, 
    loc="lower right",
    title = "Platform"
)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors=["grey", "#1aaa10"],)

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize=18,
    pad=15,
    loc="center"
)

for autotext in autotexts:
    autotext.set_fontsize(16)
    autotext.set_color("white")
    autotext.set_weight("bold")

ax.legend(
    wedges, 
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5, 0.95),
    ncol= 2
)


plt.tight_layout()
plt.show()
```

```python
# Histograms

plt.figure(figsize=(10,6))
plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=20,
    color="#2fba28",
    edgecolor = "white"
)

plt.xlabel("average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distribution of Average Stream Duration", fontsize=16, pad=20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
# Boxplots
# Middle 50% of the data

plt.boxplot(
    [spotify_df[spotify_df["Platform Type"]=="Free"]["Skip Rate (%)"],
    spotify_df[spotify_df["Platform Type"]=="Premium"]["Skip Rate (%)"]]
)

plt.xticks([1,2], ["Free", "Premium"])
plt.xlabel("Platform Type")
plt.ylabel("Skip Rate (%)")
plt.title("Skip Rate (%) by Platform Type")


plt.show()
```

```python
# Heatmaps
# Corelation Matrix

numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin=-1, vmax=1, cmap="coolwarm")
plt.colorbar(im, label="Pearson Corellatoin")

plt.xticks(range(len(corr.columns)), corr.columns, rotation = 45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```
